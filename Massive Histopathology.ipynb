```python
from keras.applications import Xception,DenseNet121,InceptionResNetV2,ResNet50,VGG16,VGG19,ResNet152,NASNetMobile
import keras
from keras.layers import Flatten, Dense, Dropout
from keras import layers

base_model = VGG16(weights="imagenet",
                              include_top=False,
                              input_shape=(50,50,3), pooling="avg")

predictions = keras.layers.Dense(14, activation='sigmoid', name='predictions')(base_model.output)

base_model = keras.Model(inputs=base_model.input, outputs=predictions)
#base_model.load_weights("ChestXNet_weights.h5")

base_model.layers.pop()

for layer in base_model.layers[:-8]:
	layer.trainable = False

new_model = keras.Model(base_model.inputs, base_model.layers[-2].output)
new_model.trainable = False

dropout1 = Dropout(0.3)
dropout2 = Dropout(0.4)
dropout3 = Dropout(0.3)
dropout4 = Dropout(0.4)
flatten = Flatten()

out = flatten(new_model.output)
out = Dense(512, activation='relu')(out)
#out = dropout1(out)
#out = Dense(128, activation='relu')(out)
#out = dropout2(out)
#out = Dense(128, activation='relu')(out)
out = dropout3(out)
out = Dense(64, activation='relu')(out)
out = dropout4(out)
output = layers.Dense(1,activation = "sigmoid",name = 'output_layer')(out)

model = keras.Model(new_model.inputs,output)

model.summary()
```

```python
from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision
    
def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
opt = optimizers.Adam(learning_rate=0.0001)
model.compile(loss='binary_crossentropy', 
                  optimizer=opt, 
                  metrics=[keras.metrics.BinaryAccuracy(),
                           keras.metrics.Recall(),
                           keras.metrics.TrueNegatives(),
                           f1_m])
```

```python
#CAM Implementation
import cv2
import numpy as np
import os
from keras import backend as kb

def create_cam(image, model):
   
    label = model.predict(image)
    index = np.where(label==np.max(label))

    img_transformed = image
    
    # CAM overlay
    # Get the 512 input weights to the softmax.
    class_weights = model.layers[-1].get_weights()[0]
    final_conv_layer = get_output_layer(model, "block4_conv3")

    get_output = kb.function([model.layers[0].input], [final_conv_layer.output, model.layers[-1].output])
    [conv_outputs, predictions] = get_output([np.array(img_transformed)])
    conv_outputs = conv_outputs[0, :, :, :]
    # Create the class activation map.
    cam = np.zeros(dtype=np.float32, shape=(conv_outputs.shape[:2]))
    for i, w in enumerate(class_weights[index]):
        cam += w * conv_outputs[:, :, i]
    # print(f"predictions: {predictions}")
    #cam = np.abs(cam)
    cam = np.abs(cam)
    cam /= np.max(cam)
    cam = cv2.resize(cam, (50,50))
    cam = np.transpose(cam,(1,0))

    #heatmap = cv2.applyColorMap(np.uint8(255 * cam), cv2.COLORMAP_JET)
    #heatmap[np.where(cam < 0.2)] = 0
    # img = heatmap * 0.5 + image[0,:,:,:]
    return cam


def get_output_layer(model, layer_name):
    # get the symbolic outputs of each "key" layer (we gave them unique names).
    layer_dict = dict([(layer.name, layer) for layer in model.layers])
    layer = layer_dict[layer_name]
    return layer
```
